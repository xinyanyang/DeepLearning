### Part 1

For part 1, a Residual Network was built based on the given architecture. For each weight layer, it contains 3 Ã— 3 filters for a specific number of input channels and output channels. The output of a sequence of ResNet basic blocks goes through a max pooling layer, and then goes to a fully-connected layer.

When loading the data, data augmentation was performed including random crop, random horizontal and normalization.

When training the model, I use the SGD optimizer with learning rate equals to 0.01.

I plot the change of the training and test loss along with the accuracy below.

```python
import pandas as pd
from matplotlib.pylab import plt 
%matplotlib inline 
from jupyterthemes import jtplot
jtplot.style(theme='onedork')
```

```python
#import the output data
plot_data = pd.read_excel("output.xlsx")
```

```python
plot_data['epoch'] = range(0, 60)
```

```python
#create the plot for loss 
#use subplots to show the plots side by side, use figuresize to set the size
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
plt.plot(plot_data['epoch'], plot_data['train_loss'], '-', color="dodgerblue", label = 'training loss')
plt.plot(plot_data['epoch'], plot_data['test_loss'], '-', color="darkorange", label = 'test loss')
plt.legend(loc='lower left')
plt.title("Training Loss & Test Loss", fontsize=16, fontweight='bold')
plt.xlabel("Number of Epoch")
plt.ylabel("Loss")

#create the plot for accuracy
plt.subplot(1, 2, 2)
plt.plot(plot_data['epoch'], plot_data['train_acc'], '--', color="dodgerblue", label = 'training accuracy')
plt.plot(plot_data['epoch'], plot_data['test_acc'], '--', color="darkorange", label = 'test accuracy')
plt.legend(loc='lower right')
plt.title("Training & Test Accuracy", fontsize=16, fontweight='bold')
plt.xlabel("Number of Epoch")
plt.ylabel("Accuracy")

plt.tight_layout()
plt.show()
```

**As we can see from this plot, as the epoch increases, the test accuracy would increase quickly in the first 20 epochs and then fluctuates around the 60%. It finally reaches at around 63% accuracy.**
